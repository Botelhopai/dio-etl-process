# EXTRACT

Extraindo dados do arquivo csv de clientes "SDW2023.csv"


```python
import pandas as pd 

df_users = pd.read_csv("SDW2023.csv")
users = df_users['Name'].tolist()
print(users)
```

# TRNANSFORM

Utilizando a API do OpenAI GPT-4 para gerar uma mensagem de marketing personalizada para cada usuário.

```python
from groq import Groq
from dotenv import load_dotenv
import os

load_dotenv()
def generate_ai_news(user):
    client = Groq(
    api_key= os.getenv("APIKEY"),
    )
    response = client.chat.completions.create(
        messages=[
            {
                "role": "system",
                "content": "Você é um especialista em markting bancário."
            },
            {
                "role": "user",
                "content": f"Crie uma mensagem para {user} sobre a importância dos investimentos (máximo de 100 caracteres)"
            }
        ],
        model="openai/gpt-oss-120b"
    )
    return response.choices[0].message.content

news_list = []

for user in users:
    news = generate_ai_news(user)
    news_list.append(news)
    
print(news_list)
```

# LOAD

```python
df_users["News"] = news_list

df_users.to_csv("Tabela_Clientes_Marketing.csv")
```
